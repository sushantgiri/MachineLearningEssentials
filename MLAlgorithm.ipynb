```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
titanic_dataset = pd.read_csv('titanic.csv')
```

```python
titanic_dataset.head(5)
```

```python
titanic_dataset.isnull().sum()
```

```python
## Continuous Variable
## Age has missing values so we should replace it with median values
titanic_dataset['Age'].head(10)
```

```python
titanic_dataset['Age'].fillna(titanic_dataset['Age'].mean(), inplace=True)
```

```python
titanic_dataset.head(10)
```

```python
## Combing SibSp and Parch
for index, col in enumerate(['SibSp','Parch']):
    plt.figure(index)
    sns.catplot(x=col, y='Survived', data=titanic_dataset, kind='point', aspect=2)
```

```python
titanic_dataset['FamilyCount'] = titanic_dataset['SibSp'] + titanic_dataset['Parch']
```

```python
titanic_dataset.head(10)
```

```python
titanic_dataset.drop(['SibSp','Parch'], axis=1)
```

```python
titanic_dataset.isnull().sum()
```

```python
titanic_dataset.groupby(titanic_dataset['Cabin'].isnull())['Survived'].mean()
```

```python
titanic_dataset['CabinInd'] = np.where(titanic_dataset['Cabin'].isnull(), 0, 1)
```

```python
titanic_dataset.head(10)
```

```python
## Converting sex to numeric
gend ={'male': 0, 'female': 1}
titanic_dataset['Sex'] = titanic_dataset['Sex'].map(gend)
```

```python
titanic_dataset
```

```python
## Dropping unnecessary columns
titanic_dataset.drop(['Cabin','Embarked','Name','PassengerId'], axis=1, inplace=True)
```

```python
titanic_dataset.head(5)
```

```python
## Writing out the cleaned data
titanic_dataset.to_csv('../cleaned_titanic_dataset.csv')
```
